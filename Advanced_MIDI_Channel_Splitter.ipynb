```
#@title Split Channels.
input_files_extension = "*.mid" #@param ["*.mid", "*.midi", "*.kar", "*.*"]
output_extension = ".mid" #@param {type:"string"}
output_files_names_tag = "_split_.mid" #@param {type:"string"}

split_channel0 = True #@param {type:"boolean"}
split_channel1 = False #@param {type:"boolean"}
split_channel2 = False #@param {type:"boolean"}
split_channel3 = False #@param {type:"boolean"}
split_channel4 = False #@param {type:"boolean"}
split_channel5 = False #@param {type:"boolean"}
split_channel6 = False #@param {type:"boolean"}
split_channel7 = False #@param {type:"boolean"}
split_channel8 = False #@param {type:"boolean"}
split_channel9 = False #@param {type:"boolean"}
split_channel10 = False #@param {type:"boolean"}
split_channel11 = False #@param {type:"boolean"}
split_channel12 = False #@param {type:"boolean"}
split_channel13 = False #@param {type:"boolean"}
split_channel14 = False #@param {type:"boolean"}
split_channel15 = False #@param {type:"boolean"}

sf = 0

debug = False #@param {type:"boolean"}

if not os.path.exists(dest_path):
    os.makedirs(dest_path)

# Create a list of paths and files. Select desired MIDI extension
source_files = os.listdir(source_path)
if debug: print(source_files)

# Define reducer/parser function and input desired instrument to reduce to { run: "auto", display-mode: "form" }

def MIDI_Splitter(fname):

    midi_file = open(fname, 'rb')
    opus = MIDI.midi2opus(midi_file.read())
    opus_in = MIDI.to_millisecs(opus)
    
    split_files_counter = 0

    channel0 = MIDI.grep(opus_in, [0])
    channel1 = MIDI.grep(opus_in, [1])
    channel2 = MIDI.grep(opus_in, [2])
    channel3 = MIDI.grep(opus_in, [3])
    channel4 = MIDI.grep(opus_in, [4])
    channel5 = MIDI.grep(opus_in, [5])
    channel6 = MIDI.grep(opus_in, [6])
    channel7 = MIDI.grep(opus_in, [7])
    channel8 = MIDI.grep(opus_in, [8])
    channel9 = MIDI.grep(opus_in, [9])
    channel10 = MIDI.grep(opus_in, [10])
    channel11 = MIDI.grep(opus_in, [11])
    channel12 = MIDI.grep(opus_in, [12])
    channel13 = MIDI.grep(opus_in, [13])
    channel14 = MIDI.grep(opus_in, [14])
    channel15 = MIDI.grep(opus_in, [15])

    if split_channel0:
      opus_out = channel0
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_0.mid').replace(source_path, dest_path)
      out_file = MIDI.opus2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1

    if split_channel1:
      opus_out = channel1
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_1.mid').replace(source_path, dest_path)
      out_file = MIDI.score2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1

    if split_channel2:
      opus_out = channel2
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_2.mid').replace(source_path, dest_path)
      out_file = MIDI.opus2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1

    if split_channel3:      
      opus_out = channel3
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_3.mid').replace(source_path, dest_path)
      out_file = MIDI.score2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1
    
    if split_channel4:      
      opus_out = channel4
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_4.mid').replace(source_path, dest_path)
      out_file = MIDI.opus2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1

    if split_channel5:  
      opus_out = channel5
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_5.mid').replace(source_path, dest_path)
      out_file = MIDI.score2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1
      
    if split_channel6:
      opus_out = channel6
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_6.mid').replace(source_path, dest_path)
      out_file = MIDI.opus2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1
      
    if split_channel7:
      opus_out = channel7
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_7.mid').replace(source_path, dest_path)
      out_file = MIDI.score2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1

    if split_channel8:
      opus_out = channel8
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_8.mid').replace(source_path, dest_path)
      out_file = MIDI.opus2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1
    
    if split_channel9:
      opus_out = channel9
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_9.mid').replace(source_path, dest_path)
      out_file = MIDI.score2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1
    
    if split_channel10:
      opus_out = channel10
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_10.mid').replace(source_path, dest_path)
      out_file = MIDI.opus2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1

    if split_channel11:  
      opus_out = channel11
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_11.mid').replace(source_path, dest_path)
      out_file = MIDI.score2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1
    
    if split_channel12:  
      opus_out = channel12
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_12.mid').replace(source_path, dest_path)
      out_file = MIDI.opus2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1
    
    if split_channel13:  
      opus_out = channel13
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_13.mid').replace(source_path, dest_path)
      out_file = MIDI.score2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1

    if split_channel14:  
      opus_out = channel14
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_14.mid').replace(source_path, dest_path)
      out_file = MIDI.opus2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1

    if split_channel15:  
      opus_out = channel15
      outname = fname.replace(output_extension, output_files_names_tag).replace(output_extension, 'MIDI_Channel_15.mid').replace(source_path, dest_path)
      out_file = MIDI.score2midi(opus_out)
      with open(outname, 'wb') as midi_file_out:
          midi_file_out.write(out_file)
          midi_file_out.close()
          split_files_counter += 1

    return split_files_counter

print('Splitting...')

for fname in tqdm.auto.tqdm(source_files):
  try:
    fname = os.path.join(source_path, fname)
    s = MIDI_Splitter(fname)
    sf += s
  except KeyboardInterrupt:
    break
  except:
    break

print('Split', len(source_files), 'MIDI files into', sf, 'single MIDI Channel files.')
print('Validated total of', len(os.listdir(dest_path)), 'MIDIs')
print('Enjoy :)')
```
